WORLD UNIVERSITY RANKINGS

THE (Times Higher Education) has been providing trusted performance data on universities for students and their families, academics, university leaders, governments and industry, since 2004. They create university rankings to assess university performance on the global stage and to provide a resource for readers to understand the different missions and successes of higher education institutions.

Our rankings cover the three main areas of university activity: research, impact and teaching.

Dataset inlcudes the below columns:

1.world_rank
2.university_name
3.country
4.teaching
5.international
6.research
7.citations
8.income
9.total_score
10.num_students
11.student_staff_ratio
12.international_students
13.female_male_ratio
14.year




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("timesData.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Plotting histogram for total_score for the year 2014
year_2014_data = df[df['year'] == 2014]
total_score_2014 = year_2014_data['total_score']
# Plotting the histogram
total_score_2014.plot(kind='hist',figsize=(10,6),edgecolor='black')

# Adding Title,x-label & y-label
plt.title('Total score for university in 2014')
plt.xlabel('Total Score')
plt.ylabel('Count')
plt.show()
```

```python
#Calcuate the frequency count and bin edges
count,bin_edges = np.histogram(total_score_2014)

print("count= ",count) # frequency count
print("bin edges= ",bin_edges) # bin ranges
```

```python
#Re-Plotting the above histogram so that the x-axis values matches with the bin size we got in the prev step
#x-axis values can be changed by adding the keyword 'xticks'
total_score_2014.plot(kind='hist',figsize=(10,6),edgecolor='black',color='teal',xticks=bin_edges)

# Adding Title,x-label & y-label
plt.title('Total score for university in 2014')
plt.xlabel('Total Score')
plt.ylabel('Count')
plt.show()
```

```python
#Calculating Student-Staff Ratio for the year 2011 and 2012
year_2011_data = df[df['year'] == 2011]
student_staff_ratio_2011 = year_2011_data['student_staff_ratio']

year_2012_data = df[df['year'] == 2012]
student_staff_ratio_2012 = year_2012_data['student_staff_ratio']
```

```python
#Increase the size of the figure
plt.figure(figsize=(12, 6))
# Plotting histograms for 2011 and 2012
student_staff_ratio_2011.plot( kind='hist' ,bins=20, alpha=0.4, label='2011', color='orange', edgecolor='black')
student_staff_ratio_2012.plot(kind='hist',bins=20, alpha=0.4, label='2012', color='skyblue', edgecolor='black')
# Adding labels and title
plt.title('Student-Staff Ratio Comparison (2011 and 2012)')
plt.xlabel('Student-Staff Ratio')
plt.ylabel('Frequency')

# Adding legend

plt.legend()

# Display the plot
plt.show()
```

```python
#Increase the size of the figure
plt.figure(figsize=(12, 6))
# Plotting histograms for 2011 and 2012
student_staff_ratio_2012.plot( kind='hist', bins=20,label='2012', color='skyblue',stacked=True, edgecolor='black')
student_staff_ratio_2011.plot( kind='hist', bins=20,label='2011', color='lightyellow', stacked=True,edgecolor='black')


# Adding labels and title
plt.title('Student-Staff Ratio Comparison (2011 and 2012)')
plt.xlabel('Student-Staff Ratio')
plt.ylabel('Frequency')

# Adding legend
plt.legend()


# Display the plot
plt.show()
```
